Question 1: Crie uma função que recebe uma lista de números inteiros e retorne uma nova lista com apenas os números primos em ordem crescente e sem repetições. Caso a lista não possua números primos, deve-se apresentar uma mensagem que avise ao usuário que a lista não possui números primos. Considerando as listas `lista_1 = [12, 367, 541, 270, 875, 14, 440, 979, 732]` e `lista_2 = [378, 928, 533, 851, 470]`, qual das alternativas apresenta os retornos corretos da função criada para cada uma das listas, respectivamente?

Escolha uma opção:

a. `[367, 541, 979] e [851]`, respectivamente

b. `[367, 541]` e `[533, 851]`, respectivamente

c. A lista não possui números primos!!! e `[367, 541]`, respectivamente

**d. `[367, 541]` e A lista não possui números primos!!!, respectivamente**

```python
from sympy import *

def prime_numbers(list3):

  list_prime_number = []

  for number in list3:
    if isprime(number) and number not in list_prime_number:
      list_prime_number.append(number)

  return list_prime_number

def print_message(list4):
  if len(list4) == 0:
    print('A lista não possui números primos!!!')
  else:
    print(list4)

l1 = [12, 367, 541, 270, 875, 14, 440, 979, 732]
l2 = [378, 928, 533, 851, 470]

l1_prime_numbers = prime_numbers(l1)
l2_prime_numbers = prime_numbers(l2)

print_message(l1_prime_numbers)
print_message(l2_prime_numbers)
```

Substitua todos os valores ímpares da matriz por -1 e calcule a soma de todos os elementos da matriz modificada. Quais das alternativas apresentam o valor da soma calculada?

Escolha uma opção:

a. 594

b. 371

**c. 1078**

d. 543

```python
import numpy as np

matrix = np.array([[17, 12, 67, 72, 64, 80, 84], [55, 48, 3, 87, 82, 60, 73], [97, 25, 1, 28, 28, 14, 56],
 [99, 5, 26, 59, 91, 8, 29], [81, 41, 8, 71, 47, 51, 32], [59, 90, 65, 67, 65, 26, 94], [53, 56, 4, 60, 9, 11, 72]])

# matrix.ndim

with np.nditer(matrix, op_flags=['readwrite']) as it:
  for number in it:
    if number%2 != 0:
      number[...] = -1

column_sum = matrix.sum(axis=1).sum(0)
print(column_sum)
```

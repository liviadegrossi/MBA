```python
import pandas as pd

df = pd.read_csv('abalone.csv')
df.columns
```

```python
# Calculating the mean of both 'Diameter' and 'Whole weight' parameters

mean_parameters = df[['Diameter', 'Whole weight']].mean()
print(mean_parameters)
```

```python
# Calculating the variance of both 'Diameter' and 'Whole weight' parameters
variance_parameters = df[['Diameter', 'Whole weight']].var()
print(variance_parameters)
```

```python
# Calculating the skewness  of both 'Diameter' and 'Whole weight' parameters
skewness_parameters = df[['Diameter', 'Whole weight']].skew()
print(skewness_parameters)
```

```python
# Calculating the kurtosis of both 'Diameter' and 'Whole weight' parameters
kurtosis_parameters = df[['Diameter', 'Whole weight']].kurtosis()
print(kurtosis_parameters)
```

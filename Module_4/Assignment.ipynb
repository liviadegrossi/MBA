```python
import findspark 
findspark.init() # makes pyspark importable as a regular library
```

```python
from pyspark import SparkContext
```

```python
# Initializing the SparkContext
sc = SparkContext("local", "Assignment")
```

```python
# Creating an RDD of integers
rdd = sc.parallelize([1, 2, 3, 4, 5])

transformed_rdd = rdd.map(lambda x: x * 3)
filtered_rdd = transformed_rdd.filter(lambda x: x > 8)
```

```python
# Collecting the results
results = filtered_rdd.collect()
print(results)
```

```python
# Stop the SparkContext 
sc.stop()
```
